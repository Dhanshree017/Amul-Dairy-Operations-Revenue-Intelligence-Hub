```python
import pandas as pd
import numpy as np

# 1. Load the dataset
df = pd.read_csv("/content/train.csv")

# 2. Clean column names (remove any accidental spaces)
df.columns = df.columns.str.strip()

# 3. Convert all Date columns to proper datetime format
date_cols = ['Date', 'Production Date', 'Expiration Date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# 4. Handle Missing Values safely
# Fill missing stock thresholds with the median value
if 'Minimum Stock Threshold (liters/kg)' in df.columns:
    median_threshold = df['Minimum Stock Threshold (liters/kg)'].median()
    df['Minimum Stock Threshold (liters/kg)'].fillna(median_threshold, inplace=True)

# 5. Create a clean 'Total Revenue' check column
if 'Quantity Sold (liters/kg)' in df.columns and 'Price per Unit (sold)' in df.columns:
    df['Calculated_Revenue'] = df['Quantity Sold (liters/kg)'] * df['Price per Unit (sold)']

# 6. Save the perfectly cleaned data
df.to_csv("Amul_Factory_Cleaned.csv", index=False)
print("Data cleaning complete! Amul_Factory_Cleaned.csv is ready for download.")
```
